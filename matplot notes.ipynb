```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

### Line Plots

```python
# Creating a single plot

# x-values
x_axis = np.arange(0, 5, 0.1)
# y-values
e_x = [np.exp(x) for x in x_axis]

# Create and show graph
plt.plot(x_axis, e_x)

# Axis Labels
plt.xlabel("Time With MatPlotLib")
plt.ylabel("How Cool MatPlotLib Seems")

plt.show()
```

```python
# Creating multiple plots on same graph

# x-values
x_axis = np.arange(0, 10, 0.1)
# sin y-values
sin = np.sin(x_axis)
# cos y-values
cos = np.cos(x_axis)

plt.plot(x_axis, sin)
plt.plot(x_axis, cos)

plt.show()
```

```python
# Creates horizontal line
# hlines(y=0, from 0, to 10, 25% transparency)
plt.hlines(0, 0, 10, alpha=0.25)

# Assign plots to tuples that stores result of plot
# format each curve
sine_handle, = plt.plot(x_axis, sin, marker ='o', color='blue', label="Sine")
cosine_handle, = plt.plot(x_axis, cos, marker='^', color='red', label="Cosine")

# Adds a legend
plt.legend(loc="lower right")

# Exports plot as image
# plt.savefig("../Images/lineConfig.png")
plt.show()
```


### Bar Charts

```python
# x,y data
users = [13000, 26000, 52000, 30000, 9000]
x_axis = np.arange(len(users))
labels = ["Java", "C++", "Python", "Ruby", "Clojure"]

# Create bar plot
plt.bar(x_axis, users, color='r', alpha=0.5)

# Where to place each x-axis headers
tick_locations = [value for value in x_axis]

# Name x-axis plots
plt.xticks(tick_locations, labels)

# Sets the x, y limits
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(users)+5000)

# Give our chart some labels and a tile
plt.title("Popularity of Programming Languages")
plt.xlabel("Programming Language")
plt.ylabel("Number of People Using Programming Languages")
plt.show()
```

### Pie Charts

```python
# Main data
labels = ["Humans", "Smurfs", "Hobbits", "Ninjas"]
sizes = [220, 95, 80, 100]

# Set desired colors
colors = ["red", "orange", "lightcoral", "lightskyblue"]

# Separate the "Humans" section
explode = (0.1, 0, 0, 0)

# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.2f%%", shadow=True, startangle=120)

# Make perfect circle (Shouldn't matter)
plt.axis("equal")
plt.show()
```

### Scatter Plots

```python
# Set the maximum x value
x_limit = 100

# x, y data
x_axis = np.arange(0, x_limit, 1)
data = [random.random() for value in x_axis]

# Create Scatterplot
# The size of each point on our plot is determined by their x value
# set s=integer for specific size
plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
            s=x_axis, alpha=0.5)


# Create limits
plt.ylim(0, 1)
plt.xlim(0, x_limit)

plt.show()
```

# Plotting with Pandas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Bar Graphs

```python
# Load in csv
rain_df = pd.read_csv("/Users/ethandonoho/Desktop/bootcamp/UofM-VIRT-DATA-PT-09-2024-U-LOLC/05-Data-Visualization/2/Activities/02-Ins_PandasPlot/Resources/avg_rain_state.csv")
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
tick_locations = [value for value in x_axis]

# set graph to 20'x4'
plt.figure(figsize=(20,4))
plt.bar(x_axis, rain_df["Inches"], color='r', alpha=0.5, align="center")
# sets location of each x-value
# labels are vertical
plt.xticks(tick_locations, rain_df["State"], rotation="vertical")

# Set x and y graph limits
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(rain_df["Inches"])+10)

# Set a Title and labels
plt.title("Average Rain per State")
plt.xlabel("State")
plt.ylabel("Average Amount of Rainfall in Inches")

# Save our graph and show the graph
# plt.tight_layout()
# plt.savefig("../Images/avg_state_rain.png")
plt.show()
```

```python
# Filter the DataFrame down only to those columns to chart
state_and_inches = rain_df[["State","Inches"]]

# Set the index to be "State" so they will be used as labels
state_and_inches = state_and_inches.set_index("State")

state_and_inches.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
state_and_inches.plot.bar(figsize=(20,3.5))

# Set a title for the chart
plt.title("Average Rain Per State")

plt.tight_layout()
plt.show()
```

```python
# Plotting mutliple on same graph

# Pandas can also plot multiple columns if the DataFrame includes them
multi_plot = rain_df.plot(kind="bar", figsize=(20,5))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(rain_df["State"], rotation=45)

plt.tight_layout()
plt.show()
```

### From Count

```python
# Import our data into pandas from CSV
accident_string = '/Users/ethandonoho/Desktop/bootcamp/UofM-VIRT-DATA-PT-09-2024-U-LOLC/05-Data-Visualization/2/Activities/04-Ins_GroupPlots/Resources/accidents.csv'
accidents_df = pd.read_csv(accident_string, low_memory=False)

accidents_df
```

```python
# Create a group based on the values in the 'FUNC_SYSNAME' column
accident_road_type = accidents_df.groupby('FUNC_SYSNAME')

# Count how many times each road type appears in our group
count_road_types = accident_road_type['FUNC_SYSNAME'].count()

count_road_types
```

```python
# Create a bar chart based off of the group series from before
count_chart = count_road_types.plot(kind='bar', figsize=(6,8))

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Road Type")
count_chart.set_ylabel("Number of Accidents")

plt.tight_layout()
plt.show()
```
